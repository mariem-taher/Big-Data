```python
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# supprimer les espaces (au cas o ils existent)
energy_data.columns = energy_data.columns.str.strip()

energy_data['Timestamp'] = pd.to_datetime(energy_data['Timestamp'])
energy_data['Humidity'] = pd.to_numeric(energy_data['Humidity'], errors='coerce')
energy_data['EnergyConsumption'] = pd.to_numeric(energy_data['EnergyConsumption'], errors='coerce')

# Remplir les valeurs manquantes avec la mdiane
imputer = SimpleImputer(strategy='median')
energy_data[['Humidity', 'EnergyConsumption']] = imputer.fit_transform(energy_data[['Humidity', 'EnergyConsumption']])

# Utiliser'Timestamp' comme index:
energy_data.set_index('Timestamp', inplace=True)

# Normaliser(ou bien standardiser) les donnes :
scaler = StandardScaler()
scaled_features = scaler.fit_transform(energy_data[['Humidity', 'EnergyConsumption']])

kmeans = KMeans(n_clusters=3, random_state=42)
energy_data['cluster'] = kmeans.fit_predict(scaled_features)

# on calcule les moyennes pour chaque cluster:
cluster_means = energy_data.groupby('cluster')[['Humidity', 'EnergyConsumption']].mean()
forecast_hours = 24 * 7
forecast_dates = pd.date_range(start=energy_data.index[-1], periods=forecast_hours+1, freq='H')[1:]

# stocker les prvisions dans forecast_data
forecast_data = pd.DataFrame(index=forecast_dates, columns=['Humidity', 'EnergyConsumption'])

# on va choisir un cluster alatoire pour simuler les futurs comportements
chosen_cluster = np.random.choice(energy_data['cluster'].unique())

# on extrait les donnes dans recent_data:
recent_data = energy_data[energy_data['cluster'] == chosen_cluster].tail(24)

# Simuler les variations bases sur les diffrences:
delta_humidity = recent_data['Humidity'].diff().dropna()  # Variation de l'humidit
delta_energy = recent_data['EnergyConsumption'].diff().dropna()  # Variation de la consommation

# Rpter les variations pour les appliquer sur la priode de prvision
future_humidity = np.tile(delta_humidity.values, forecast_hours // len(delta_humidity) + 1)[:forecast_hours]
future_energy = np.tile(delta_energy.values, forecast_hours // len(delta_energy) + 1)[:forecast_hours]

# Les dernires valeurs actuelles d'humidit et de consommation
start_humidity = recent_data['Humidity'].iloc[-1]
start_energy = recent_data['EnergyConsumption'].iloc[-1]

# on applique les changements  partir des dernires valeurs
forecast_data['Humidity'] = start_humidity + np.cumsum(future_humidity)
forecast_data['EnergyConsumption'] = start_energy + np.cumsum(future_energy)

# Humidit : Donnes actuelles VS prvisions
plt.figure(figsize=(30, 12))
plt.plot(energy_data.index[-24*7:], energy_data['Humidity'][-24*7:], label='Actual Humidity', color='red', linewidth=3)
plt.plot(forecast_data.index, forecast_data['Humidity'], label='Predicted Hulidity', color='orange', linewidth=3)
plt.title("Humidity Prediction for Next Week", fontsize=30)
plt.xlabel('Time (Hours)', fontsize=30)
plt.ylabel('Humidity(%)', fontsize=30)
plt.legend(fontsize=20)
plt.xticks(rotation=45)
plt.show()

# Consommation d'nergie : Donnes actuelles VS prvisions
plt.figure(figsize=(30, 12))
plt.plot(energy_data.index[-24*7:], energy_data['EnergyConsumption'][-24*7:], label='Actual Consumption', color='blue', linewidth=3)
plt.plot(forecast_data.index, forecast_data['EnergyConsumption'], label='Predicted Consumption', color='green', linewidth=3)
plt.title("EnergyConsumption Prediction for Next Week", fontsize=30)
plt.xlabel('Time (Hours)', fontsize=30)
plt.ylabel('Energy Consumption(kW)', fontsize=30)
plt.legend(fontsize=20)
plt.xticks(rotation=45)
plt.show()
```

```python
energy_dt=energy_data.copy()
energy_dt = energy_dt.reset_index()
energy_dt['Timestamp'] = pd.to_datetime(energy_dt['Timestamp'], errors='coerce')
energy_dt['DayOfWeek'] = energy_dt['Timestamp'].dt.dayofweek
energy_dt['Hour'] = energy_dt['Timestamp'].dt.hour
energy_dt['Day'] = energy_dt['Timestamp'].dt.day
energy_dt['weekend']= energy_dt['DayOfWeek'].apply(lambda x: 1 if x == 6 else 0)
total_energy_consumption = energy_dt.groupby('weekend')['EnergyConsumption'].agg(['sum', 'mean', 'median', 'std'])
Total_energy_per_holiday=energy_dt.groupby('Holiday')['EnergyConsumption'].agg(['sum', 'mean', 'median', 'std'])
Total_energy_per_LightingUsage=energy_dt.groupby('LightingUsage')['EnergyConsumption'].agg(['sum', 'mean', 'median', 'std'])
print("total Energy Consumption_using weekend feature",total_energy_consumption,'\n')
print("total Energy Consumption_using holiday feature",Total_energy_per_holiday,'\n')
print("total Energy Consumption_LightingUsage feature",Total_energy_per_LightingUsage,'\n')
```

```python
energy_dt = energy_dt.drop(columns=['Timestamp'])
energy_dt.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the 'day' column
energy_dt['HVACUsage'] = label_encoder.fit_transform(energy_dt['HVACUsage'])
energy_dt['LightingUsage'] = label_encoder.fit_transform(energy_dt['LightingUsage'])
energy_dt['Holiday'] = label_encoder.fit_transform(energy_dt['Holiday'])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(energy_dt)

X = energy_dt.drop(columns=['EnergyConsumption'])
y = energy_dt['EnergyConsumption']

# split into train test sets
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)

model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue')

plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')

plt.xlabel('Actual Values (y_test)')
plt.ylabel('Predicted Values (y_pred)')
plt.title('Actual vs Predicted Values')
plt.show()
```
